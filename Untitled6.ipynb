```python
import pandas as pd
import matplotlib.pyplot as plt


df  =pd.read_excel('/content/sample_data/Задание для Аналитика_рус_24.06.20.xlsx')
new = df.iloc[0]
df = df[1:]
df.columns = new
df
```

```python
data = df[['Регион','Тип местности','Пол респондента']]

def count_respondents(data, category):
    counts = {}
    total = len(data)

    for index, row in data.iterrows():
        value = row[category]
        if value in counts:
            counts[value] += 1
        else:
            counts[value] = 1

    percentages = {key: (count / total) * 100 for key, count in counts.items()}
    return counts, percentages
# Подсчет количества и доли респондентов по регионам
region_counts, region_percentages = count_respondents(data, "Регион")

# Подсчет количества и доли респондентов по типу местности
locality_counts, locality_percentages = count_respondents(data, "Тип местности")

# Подсчет количества и доли респондентов по полу
gender_counts, gender_percentages = count_respondents(data, "Пол респондента")

# Построение диаграммы для регионов
plt.figure(figsize=(8, 6))
plt.pie(region_counts.values(), labels=region_counts.keys(), autopct='%1.1f%%')
plt.title("Доля респондентов по регионам")
plt.show()

# Построение диаграммы для типа местности
plt.figure(figsize=(8, 6))
plt.pie(locality_counts.values(), labels=locality_counts.keys(), autopct='%1.1f%%')
plt.title("Доля респондентов по типу местности")
plt.show()

# Построение диаграммы для пола
plt.figure(figsize=(8, 6))
plt.pie(gender_counts.values(), labels=gender_counts.keys(), autopct='%1.1f%%')
plt.title("Доля респондентов по полу")
plt.show()
```

```python
# Предположим, что у вас есть DataFrame с данными
df[['компьютер', 'мобильный телефон', 'VPN']] = df[['Скажите, пожалуйста, как часто в течение последних 6 месяцев Вы выходили в Интернет используя компьютер?','Скажите, пожалуйста, как часто в течение последних 6 месяцев Вы выходили в Интернет используя сотовый телефон?','Скажите, пожалуйста, как часто в течение последних 6 месяцев Вы выходили в Интернет через VPN?']]
df['компьютер'] = df['компьютер'].map({
    'Никогда': 0,
    'Несколько раз в месяц': 1,
    'Редко': 2,
    'Не знаю [НЕ ЗАЧИТЫВАТЬ]':3,
    'Ежедневно': 4
})
df['мобильный телефон'] = df['мобильный телефон'].map({
    'Никогда': 0,
    'Несколько раз в месяц': 1,
    'Редко': 2,
    'Не знаю [НЕ ЗАЧИТЫВАТЬ]':3,
    'Ежедневно': 4
})
df['VPN'] = df['VPN'].map({
    'Никогда': 0,
    'Несколько раз в месяц': 1,
    'Редко': 2,
    'Не знаю [НЕ ЗАЧИТЫВАТЬ]':3,
    'Ежедневно': 4
})
data = df

# Группировка данных по регионам и вычисление долей
region_data = data.groupby('Регион').agg({
    'компьютер': 'mean',
    'мобильный телефон': 'mean',
    'VPN': 'mean'
})

# Группировка данных по типу местности и вычисление долей
location_data = data.groupby('Тип местности').agg({
    'компьютер': 'mean',
    'мобильный телефон': 'mean',
    'VPN': 'mean'
})

# Группировка данных по полу респондентов и вычисление долей
gender_data = data.groupby('Пол респондента').agg({
    'компьютер': 'mean',
    'мобильный телефон': 'mean',
    'VPN': 'mean'
})

# Создание таблиц и диаграмм для регионов
region_data['Всего респондентов'] = data['Регион'].value_counts()
region_data.plot(kind='bar', y=['компьютер', 'мобильный телефон', 'VPN'], stacked=True)
plt.title('Доли респондентов по регионам')
plt.xlabel('Регион')
plt.ylabel('Доля респондентов')
plt.show()

# Создание таблиц и диаграмм для типов местности
location_data['Всего респондентов'] = data['Тип местности'].value_counts()
location_data.plot(kind='bar', y=['компьютер', 'мобильный телефон', 'VPN'], stacked=True)
plt.title('Доли респондентов по типам местности')
plt.xlabel('Тип местности')
plt.ylabel('Доля респондентов')
plt.show()

# Создание таблиц и диаграмм для пола респондентов
gender_data['Всего респондентов'] = data['Пол респондента'].value_counts()
gender_data.plot(kind='bar', y=['компьютер', 'мобильный телефон', 'VPN'], stacked=True)
plt.title('Доли респондентов по полу')
plt.xlabel('Пол респондента')
plt.ylabel('Доля респондентов')
plt.show()
```

```python
df[['Сотовый телефон','Компьютер','Спутниковое телевидение','Автомобиль']] = df[['У вас есть сотовый телефон?','У вас есть компьютер?','У вас есть спутниковое телевидение?','Есть ли у Вас, либо у кого-либо из членов Вашей семьи автомобиль?']]

# Определение доли респондентов с указанными предметами
total_respondents = len(df)
cell_phone_ratio = df['Сотовый телефон'].value_counts(normalize=True)
computer_ratio = df['Компьютер'].value_counts(normalize=True)
satellite_tv_ratio = df['Спутниковое телевидение'].value_counts(normalize=True)
car_ratio = df['Автомобиль'].value_counts(normalize=True)

# Группировка данных по регионам
region_respondents = df.groupby('Регион').size().reset_index(name='Количество респондентов')
region_cell_phone_ratio = df.groupby('Регион')['Сотовый телефон'].value_counts(normalize=True).unstack().reset_index()
region_computer_ratio = df.groupby('Регион')['Компьютер'].value_counts(normalize=True).unstack().reset_index()
region_satellite_tv_ratio = df.groupby('Регион')['Спутниковое телевидение'].value_counts(normalize=True).unstack().reset_index()
region_car_ratio = df.groupby('Регион')['Автомобиль'].value_counts(normalize=True).unstack().reset_index()

# Построение линейчатых диаграмм

# Доли в целом по стране
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Доли в целом по стране')

cell_phone_ratio.plot(kind='bar', ax=axes[0, 0], rot=0)
axes[0, 0].set_title('Сотовый телефон')
axes[0, 0].set_ylabel('Доля респондентов')

computer_ratio.plot(kind='bar', ax=axes[0, 1], rot=0)
axes[0, 1].set_title('Компьютер')
axes[0, 1].set_ylabel('Доля респондентов')

satellite_tv_ratio.plot(kind='bar', ax=axes[1, 0], rot=0)
axes[1, 0].set_title('Спутниковое телевидение')
axes[1, 0].set_ylabel('Доля респондентов')

car_ratio.plot(kind='bar', ax=axes[1, 1], rot=0)
axes[1, 1].set_title('Автомобиль')
axes[1, 1].set_ylabel('Доля респондентов')

plt.tight_layout()
plt.show()

# Доли по регионам
fig, axes = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Доли по регионам')

region_cell_phone_ratio.plot(x='Регион', kind='bar', ax=axes[0, 0], rot=0)
axes[0, 0].set_title('Сотовый телефон')
axes[0, 0].set_ylabel('Доля респондентов')

region_computer_ratio.plot(x='Регион', kind='bar', ax=axes[0, 1], rot=0)
axes[0, 1].set_title('Компьютер')
axes[0, 1].set_ylabel('Доля респондентов')

region_satellite_tv_ratio.plot(x='Регион', kind='bar', ax=axes[1, 0], rot=0)
axes[1, 0].set_title('Спутниковое телевидение')
axes[1, 0].set_ylabel('Доля респондентов')

region_car_ratio.plot(x='Регион', kind='bar', ax=axes[1, 1], rot=0)
axes[1, 1].set_title('Автомобиль')
axes[1, 1].set_ylabel('Доля респондентов')

plt.tight_layout()
plt.show()
```

```python
df[['Пол','Возраст']] = df[['Пол респондента', 'Сколько вам лет?']]
data = df

bins = [18, 25, 31, 36, 41, 46, 51, 56, 61, 66, 150]
labels = ['18-24', '25-30', '31-35', '36-40', '41-45', '46-50', '51-55', '56-60', '61-65', 'Старше 65']
data['Возрастная категория'] = pd.cut(data['Возраст'], bins=bins, labels=labels, right=False)
средний_возраст = data['Возраст'].mean()
pivot_общая = pd.pivot_table(data, values='Возраст', index='Возрастная категория', aggfunc='count', fill_value=0)
pivot_общая = pivot_общая / pivot_общая.sum() * 100
pivot_пол = pd.pivot_table(data, values='Возраст', index='Возрастная категория', columns='Пол', aggfunc='size', fill_value=0)
pivot_пол = pivot_пол.div(pivot_пол.sum(axis=1), axis=0) * 100
pivot_регион = pd.pivot_table(data, values='Возраст', index='Возрастная категория', columns='Регион', aggfunc='size', fill_value=0)
pivot_регион = pivot_регион.div(pivot_регион.sum(axis=1), axis=0) * 100
print('Средний возраст респондентов:', средний_возраст)
print('\nДоля возрастных категорий по всей республике:')
print(pivot_общая)
print('\nДоля возрастных категорий по полу:')
print(pivot_пол)
print('\nДоля возрастных категорий по регионам:')
print(pivot_регион)
```
